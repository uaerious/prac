```python
# linked list
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

class LinkedList:
  def __init__(self, data):
    self.data = None

  def insert(self, data):
    node = Node(data)
    if self.head is None: # empty
      self.head = node
    else: # not empty
      node.next = self.head
      self.head = node

  def delete(self, data):
    if self.head is None: # empty
      return False # nothing to delete
    else: # not empty
      curr = self.head
      prev = self.head
      while curr and curr.data != data:
        prev = curr
        curr = curr.next
      if curr is None: # data not found
        return None
      else: # data found
        prev.next = curr.next
        return str(data) + "deleted"

  def search(self, data):
    if self.head is None: # empty
      return False
    else: #not empty
      curr = self.head
      while curr:
        if curr.data == data: # found
          return True
        else: # not found
          return False

  def count(self, data):
    if self.head is None: # empty
      return 0
    else:
      curr = self.head
      result = 0
      while curr:
        curr = curr.next
        result += 1
      return result

class Queue(LinkedList):
  def enqueue(self, data):
    node = Node(data)
    if self.head is None: # empty
      self.head = node
    else: # not empty
      curr = self.head
      prev = self.head
      while curr:
        prev = curr
        curr = curr.next
      prev.next = node

  def dequeue(self):
    if self.head is None:
      return "empty"
    else:
      data = self.head.data
      self.head = self.head.next
    return data
```

```python
# bubble sort
def bubble_sort(A):
  n = len(A)
  for i in range(n):
    for index in range(n-1):
      if A[index] > A[index+1]:
        A[index], A[index+1] = A[index+1], A[index]
  return A
```

```python
# insertion sort
def insertion_sort(A):
  for i in range(1, len(A)):
    key = A[i]
    j = i - 1
    while j > 0 and key < A[j]:
      A[j+1] = A[j]
      j -= 1
    A[j+1] = key
  return A
```

```python
# quicksort
def quicksort(A):
  if len(A) <= 1:
    return A

  pivot = A[len(A) // 2]
  left = []
  middle = []
  right = []

  for element in A:
    if element < pivot:
      left.append(element)
    elif element == pivot:
      middle.append(element)
    else:
      right.append(element)

  return quicksort(left) + middle + quicksort(right)
```

```python
# mergesort
def mergesort(A):
  if len(A) <= 1:
    return A

  mid = len(A) // 2
  left = A[:mid]
  right = A[mid:]

  left = mergesort(left)
  right = mergesort(right)

  return merge(left, right)

def merge(left, right):
  i = j = 0
  merged = []
  while i < len(left) and j < len(right):
    if left[i] < right[j]:
      merged.append(left[i])
      i += 1
    else:
      merged.append(right[j])
      j += 1

  merged += left[i:]
  merged += right[j:]

  return merged
```
